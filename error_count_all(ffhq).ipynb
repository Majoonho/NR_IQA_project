```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
error_count_gau=pd.read_csv('C:/Users/Home/Desktop/IQA/error_count_gau.csv')
error_count_lens=pd.read_csv('C:/Users/Home/Desktop/IQA/error_count_lens.csv')
error_count_motion=pd.read_csv('C:/Users/Home/Desktop/IQA/error_count_motion.csv')
```

```python
error_count_gau=error_count_gau[error_count_gau.columns[1:]]
```

```python
error_count_lens=error_count_lens[error_count_lens.columns[1:]]
```

```python
error_count_motion=error_count_motion[error_count_motion.columns[1:]]
```

```python
print('error_count_gau')
print(error_count_gau.sum().sort_values().nsmallest(5))
print('-----------------------')
print('error_count_lens')
print(error_count_lens.sum().sort_values().nsmallest(5))
print('-----------------------')
print('error_count_motion')
print(error_count_motion.sum().sort_values().nsmallest(9))      
      
      
```

```python
# blur_gau 에서는 압도적으로 HYPER 가 좋음 (motion에서는 HYPER가 순위안에 못들어감) gau-1등 lens-2등 motion-9등 꼴지..
# 그 외 대부분 비슷한 경향을 보이며, 그래도 CPBD가 3가지 blur에서 모두 상위권을 차지함
# blur_motion 에서는 BRISQUE, WaDIQaM(_TDI) 계열이 좋음

#대부분 큰 차이는 나지 않아서 비슷하게 생각됨

# gau를 위해 HYPER는 꼭 들어가고 HYPER를 보완하기 위해 CPBD, BRISQUE, WaDIQaM 을 합성하면 어떨까

```

```python
plt.figure(figsize=(13,12))
n=1
for i in error_count_gau.columns[1:]:
    
    plt.subplot(3, 3, n) #sharex=ax1)
    plt.bar(['lv2','lv3','lv4','lv5'], error_count_gau[f'{i}'],  color = ['gold', 'b', '#FF0000', 'green'], alpha=0.5, label=f'{i}')
    plt.ylabel('Sum of error', fontsize=10)
    plt.xlabel((f'{i}', error_count_gau[f'{i}'].sum()), fontsize=10)
    n+=1
```

```python
plt.figure(figsize=(13,12))
n=1
for i in error_count_lens.columns[1:]:
    
    plt.subplot(3, 3, n) #sharex=ax1)
    plt.bar(['lv2','lv3','lv4','lv5'], error_count_lens[f'{i}'],  color = ['gold', 'b', '#FF0000', 'green'], alpha=0.5, label=f'{i}')
    plt.ylabel('Sum of error', fontsize=10)
    plt.xlabel((f'{i}', error_count_lens[f'{i}'].sum()), fontsize=10)
    n+=1
```
